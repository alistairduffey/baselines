```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import xarray as xr
from xmip.preprocessing import rename_cmip6
from nc_processing import calc_spatial_mean
matplotlib.rcParams.update({'font.size': 14})


### first define function to get PI gmst:
def get_pi(model, var='tas'):
    dir_pi = glob.glob('/badc/cmip6/data/CMIP6/*/*/{m}/piControl/r1i*/Amon/{v}/*/latest/'.format(m=model, v=var))
    files_pi = os.listdir(dir_pi[0])[0:3] # don't need the full length run
    paths_pi = []
    for x in files_pi:
        paths_pi.append(dir_pi[0]+x)
    da_pi = rename_cmip6(xr.open_mfdataset(paths_pi))[var].mean(dim='x')
    da_pi = da_pi.weighted(weights=np.cos(np.deg2rad(da_pi.y))).mean(dim='y')
    
    gmst_pi = da_pi.mean(dim='time').values
    print(gmst_pi)
    return gmst_pi

### also get gmst_pi_dicts by model for land and sea only:
def get_pi_region(model, region, land_frac=None):
    if not land_frac:
        land_frac = land_frac_dict[model] 
    dir_pi = glob.glob('/badc/cmip6/data/CMIP6/*/*/{}/piControl/r1i*/Amon/tas/*/latest/'.format(model))
    files_pi = os.listdir(dir_pi[0])[0:3] # don't need the full length run
    paths_pi = []
    for x in files_pi:
        paths_pi.append(dir_pi[0]+x)
    da_pi = rename_cmip6(xr.open_mfdataset(paths_pi)).tas
    if region == 'land':
        da_pi = da_pi.where(land_frac.sftlf>99, drop=True).mean(dim='x')
    elif region == 'ocean':
        da_pi = da_pi.where(land_frac.sftlf<1, drop=True).mean(dim='x')
    da_pi = da_pi.weighted(weights=np.cos(np.deg2rad(da_pi.y))).mean(dim='y')
    
    gmst_pi = da_pi.mean(dim='time').values
    print(gmst_pi)
    return gmst_pi
```

```python
### repeat for LSR

lat_band = [-90, 90] ## DONT CHANGE
transient_base_period = [2014, 2033]

### get ensemble mean tas for both scenarios 

ds_list_tas = []
paths_tas = glob.glob('/badc/deposited2022/arise/data/ARISE/MOHC/UKESM1-0-LL/arise-sai-1p5/*/Amon/tas/gn/*/')

for path in paths_tas:
    #files = os.listdir(path)
    ds = rename_cmip6(xr.open_mfdataset(path+'*.nc'))
    ds_list_tas.append(ds)

DS_ARISE_tas = xr.concat(ds_list_tas, dim='Ensemble_member')
DS_ssp245_tas = get_ssp245_ds(variable='tas', table='Amon')

### group to get annual mean tas
DS_ssp245_tas = DS_ssp245_tas.groupby('time.year').mean('time')
DS_ARISE_tas = DS_ARISE_tas.groupby('time.year').mean('time')

land_frac_UKESM = rename_cmip6(xr.open_dataset('/badc/cmip6/data/CMIP6/CMIP/MOHC/UKESM1-0-LL/piControl/r1i1p1f2/fx/sftlf/gn/latest/sftlf_fx_UKESM1-0-LL_piControl_r1i1p1f2_gn.nc'))
gmst_pi_land, gmst_pi_ocean, gmst_pi = get_pi_region('UKESM1-0-LL', 'land', land_frac=land_frac_UKESM),  get_pi_region('UKESM1-0-LL', 'ocean', land_frac=land_frac_UKESM), get_pi('UKESM1-0-LL')
```

```python
### now get land-sea warming ratio:

def landsea_warming_ratio(ds, pi_T_land, pi_T_ocean, pi_T_all, lat_bnds):
    """ get the ratio of warming relative to PI for land regions vs ocean regions
    INPUTS:
    
    ds; ds of annual mean tas, dims: (x, y, year, Ensemble_member)
    pi_land, pi_ocean, pi_T_global; pre-industrial temps over each region
    lat_bnds; latitude region (e.g. tropics) over which to calculate land/sea ratio. -90, 90 for global. 
    
    OUTPUTS:
    ds_out; ds of ratio land warming:ocean warming, dims: (year, Ensemble_member)
    """
    da_all = calc_spatial_mean(ds.tas.sel(y=slice(lat_band[0], lat_band[1])), lon_name="x", lat_name="y") - pi_T_all
    
    da_land = ds.where(land_frac_UKESM.sftlf>99, drop=True).tas.mean(dim='x').sel(y=slice(lat_band[0], lat_band[1]))
    da_land = da_land.weighted(weights=np.cos(np.deg2rad(da_land.y))).mean(dim='y') - pi_T_land

    da_ocean = ds.where(land_frac_UKESM.sftlf<1, drop=True).tas.mean(dim='x').sel(y=slice(lat_band[0], lat_band[1]))
    da_ocean = da_ocean.weighted(weights=np.cos(np.deg2rad(da_ocean.y))).mean(dim='y') - pi_T_ocean

    da_ratio = da_land/da_ocean
    ds_out = xr.merge([da_all.rename('All_temp_anom'),
                       da_land.rename('Land_temp_anom'),
                       da_ocean.rename('Ocean_temp_anom'), 
                       (da_land-da_ocean).rename('LandSea_temp_anom_diff'),
                       da_ratio.rename('LandSea_temp_anom_ratio')])
    return ds_out

arise = landsea_warming_ratio(DS_ARISE_tas, pi_T_land=gmst_pi_land, pi_T_ocean=gmst_pi_ocean,
                      pi_T_all=gmst_pi, lat_bnds=lat_band).compute()
ssp245 = landsea_warming_ratio(DS_ssp245_tas, pi_T_land=gmst_pi_land, pi_T_ocean=gmst_pi_ocean,
                      pi_T_all=gmst_pi, lat_bnds=lat_band).compute()
```
